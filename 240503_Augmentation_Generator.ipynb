```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img, ImageDataGenerator
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load image
img = load_img('/content/drive/MyDrive/Samples/dog.jpeg')
plt.imshow(img)
```

```python
# Transform image
x = img_to_array(img)           # transform into array
x = x.reshape((1,) + x.shape)   # add batch dimension
```

```python
# Image augmentation
datagen = ImageDataGenerator(
    rotation_range=40,          # 0~40도 사이 각도로 랜덤하게 회전
    width_shift_range=0.2,      # 전체 너비의 20% 범위 내에서 랜덤하게 수평 이동
    height_shift_range=0.2,     # 전체 높이의 20% 범위 내에서 랜덤하게 수직 이동
    zoom_range=0.2,             # 80~120% 사이에서 랜덤하게 확대 or 축소
    horizontal_flip=True,       # 이미지를 수평 방향으로 뒤집기
    fill_mode='nearest'         # 이미지 변형 후 생기는 공백을 가장 가까운 픽셀 값으로 채움
)

i = 0
for batch in datagen.flow(x, batch_size=1):
    # batch_size=1: 한 번에 하나의 이미지 생성
    plt.figure(i)   # 그림창 생성 (i: 그림창 고유 index)
    imgplot = plt.imshow(array_to_img(batch[0]))
    i += 1
    if (i > 2):
        break
```
