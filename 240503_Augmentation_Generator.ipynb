```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img, ImageDataGenerator
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load image
img = load_img('/content/drive/MyDrive/Samples/dog.jpeg')
plt.imshow(img)
```

```python
# Transform image
x = img_to_array(img)           # transform into array
x = x.reshape((1,) + x.shape)   # add batch dimension
```

```python
# Image augmentation
datagen = ImageDataGenerator(
    rotation_range=40,          # 0~40    
    width_shift_range=0.2,      #   20%     
    height_shift_range=0.2,     #   20%     
    zoom_range=0.2,             # 80~120%    or 
    horizontal_flip=True,       #    
    fill_mode='nearest'         #          
)

i = 0
for batch in datagen.flow(x, batch_size=1):
    # batch_size=1:     
    plt.figure(i)   #   (i:   index)
    imgplot = plt.imshow(array_to_img(batch[0]))
    i += 1
    if (i > 2):
        break
```
