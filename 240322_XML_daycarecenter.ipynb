```python
# Load dataset
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Samples/DaycareCenter.xml'

import xml.etree.ElementTree as ET
tree = ET.parse(file_path)
root = tree.getroot()
```

```python
# Extract sidoname
sidoname = root.find('.//sidoname').text if root.find('.//sidoname') is not None else '지역명 없음'
print(f'지역명: {sidoname}')
```

```python
# Extract information and make a dataframe
import pandas as pd

data = []
for item in root.findall('.//item'):
    sidoname = item.find('sidoname').text if item.find('sidoname') is not None else '지역명 없음'
    sigunname = item.find('sigunname').text if item.find('sigunname') is not None else '시군명 없음'
    stcode = item.find('stcode').text if item.find('stcode') is not None else '기관코드 없음'
    data.append({'지역명': sidoname, '시군명': sigunname, '기관코드': stcode})

df = pd.DataFrame(data)
display(df)
```
