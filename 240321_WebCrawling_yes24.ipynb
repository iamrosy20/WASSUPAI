```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.yes24.com/Product/Goods/125101376'
r = requests.get(url)
print(r.status_code)
```

```python
html = r.text
soup = BeautifulSoup(html, 'lxml')
id = 125101376
print('ID:', id)
title = soup.select_one('#yDetailTopWrap > div.topColRgt > div.gd_infoTop > div > h2').get_text()
print('Title:', title)
author = soup.select_one('#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_auth > a:nth-child(1)').get_text()
print('Author:', author)
price = soup.select_one('#yDetailTopWrap > div.topColRgt > div.gd_infoBot > div.gd_infoTbArea > div:nth-child(4) > table > tbody > tr:nth-child(2) > td > span > em').get_text()
print('Price:', price)
publisher = soup.select_one('#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_pub > a').get_text()
print('Publisher:', publisher)
rating = soup.select_one('#spanGdRating > a > em').get_text()
print('Rating:', rating)
pdate = soup.select_one('#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_date').get_text()
print('Publication Date:', pdate)
```

```python
book_df = pd.DataFrame(columns=['ID', 'Title'])

start_id = 125101374
end_id = 125101378

for book_id in range(start_id, end_id + 1):
    url = f'http://www.yes24.com/Product/Goods/{book_id}'
    try:
        r = requests.get(url)
        if r.status_code == 200:
            html = r.text
            soup = BeautifulSoup(html, 'lxml')
            title = soup.title
            book_df = book_df.append({'ID': book_id, 'Title': title}, ignore_index=True)
        else:
            print('Error')
    except requests.RequestException as e:
        print('Error')

display(book_df)
```
